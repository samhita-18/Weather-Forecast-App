```python
import requests
from datetime import datetime
import sys


API_KEY = "c62429f270461a0a6ab90af2d7ac735e"
WEATHER_URL = "http://api.openweathermap.org/data/2.5/weather" 
FORECAST_URL = "http://api.openweathermap.org/data/2.5/forecast"

def get_weather(url,city):
    params = {"q": city, "appid": API_KEY, "units": "metric"}
    
    
    try:
        response = requests.get(url, params = params)
        data = response.json()
        return data
    except requests.exceptions.HTTPError or http_err:
        status_code = response.status_code
        
        if status_code == 404:
            print(f"\n Oops!! city {city} not found. Please check the city again")
        elif status_code == 401:
            print("\n API ERROR: Invalid API Key. Please verify your API Key.")
        else:
            print(f"\n HTTP Error occurred: {http_err} (status code:{status_code})")
            
        return None

def display_current_weather(data,city):
    temp = data['main']['temp']
    feels_like = data['main']['feels_like']
    description = data['weather'][0]['description'].capitalize()
    humidity = data['main']['humidity']
    wind_speed = data['wind']['speed']
    city = data['name']
    country = data['sys']['country']
    
    
    print("\n" +"="*45)
    print(f"       â˜€ï¸ CURRENT WEATHER: {city}, {country}")
    print("=" * 45)
    print(f"ğŸŒ¡ï¸  Temperature: {temp}Â°C (Feels like: {feels_like}Â°C)")
    print(f"ğŸŒ¤ï¸  Condition:   {description}")
    print(f"ğŸ’§ Humidity:    {humidity}%")
    print(f"ğŸ’¨ Wind Speed:   {wind_speed} m/s")
    print("-" * 45)

def display_forecast(data):
    forecast_list = data.get('list', [])
    daily_summaries = []
    processed_dates = set()
    
    for entry in forecast_list:
        timestamp_str = entry['dt_txt']
        date_obj = datetime.strptime(timestamp_str, '%Y-%m-%d %H:%M:%S')
        date_only_str = date_obj.strftime('%Y-%m-%d')
        
        if date_only_str not in processed_dates and date_obj.hour == 12:
            
            day_label = date_obj.strftime('%a, %b %d')
            temp = entry['main']['temp']
            description = entry['weather'][0]['description'].capitalize()
            
            daily_summaries.append({
                "day": day_label,
                "temp": f"{temp:.1f}",
                "description": description
            })
            processed_dates.add(date_only_str)
            
            if len(daily_summaries) >= 5:
                break
    
    print("\nğŸ“… 5-DAY FORECAST (Mid-day)")
    print("---------------------------------------------")
    print(f"{'DAY':<15} | {'TEMP (Â°C)':<10} | {'CONDITION':<15}")
    print("---------------------------------------------")
    
    for day in daily_summaries:
        print(f"{day['day']:<15} | {day['temp']:<10} | {day['description']:<15}")
        
    print("---------------------------------------------")
    
    
def main():
    print("\n------- Simple Python weather app-------")
    try:
        city = input("entet the city name: ").strip()
        if not city:
            print("Exiting")
            sys.exit(0)
        
        print("\nâ³ Fetching data...")
        current_data = get_weather(WEATHER_URL,city)
        if current_data:
            display_current_weather(current_data, city)
            
        forecast_data = get_weather(FORECAST_URL, city)
        if forecast_data:
                display_forecast(forecast_data)
        
    except KeyboardInterrupt:
        print("\n\nScript interrupted. Goodbye!")
    except Exception as e:
        print(f"\nAn unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
              
```
